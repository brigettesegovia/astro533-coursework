```python
import numpy as np
import matplotlib.pyplot as plt


''''I used copilot to help me comment and structure my code.'''
```

```python
cond_temp_C = np.ones(3)*100
cond_temp_N = np.ones(3)*150
cond_temp_O = np.ones(3)*200
cond_temp_Fe = np.ones(3)*1320
cond_temp_Si = np.ones(3)*1300
cond_temp = np.array([cond_temp_C, cond_temp_N, cond_temp_O, cond_temp_Fe, cond_temp_Si])


C = np.array([-0.613, -0.63, -.6847])
N = np.array([-.162, -.162, -.162])
O = np.array([0.102, -0.149, -0.307])
Fe = np.array([-0.8501, -0.966, -0.9896])
Si = np.array([-0.1099, -0.76, -0.9154])
elements = [C, N, O, Fe, Si]
element_names = ['C', 'N', 'O', 'Fe', 'Si']

colors = ['red', 'blue', 'green']

# Plot
plt.figure(figsize=(10, 6))
for i in range(len(elements)):
    for j in range(len(colors)):
        # Plot the data point
        plt.plot(cond_temp[i][j], elements[i][j], 'o', color=colors[j])
        
        # Add a label next to each point
        if colors[j] == 'red':
            plt.text(cond_temp[i][j] + 10, elements[i][j], element_names[i], 
                     fontsize=9, color=colors[j], verticalalignment='center')

plt.ylabel('Depletion of Element (w.r.t. Sun)')
plt.xlabel('Condensation Temperature [K]')
plt.title('Temperature vs Element Value')
plt.ylim(-4.20, 1.20)
plt.xlim(0, 1520)
plt.grid()
plt.show()
```
